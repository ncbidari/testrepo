## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
print(df.columns)
```

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the 'ConvertedComp' column data
QUERY_CONVERTED_COMP = """
SELECT ConvertedComp
FROM master
"""
converted_comp_df = pd.read_sql_query(QUERY_CONVERTED_COMP, conn)

# Plot the histogram
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.hist(converted_comp_df['ConvertedComp'].dropna(), bins=30, edgecolor='black')  # Drop NaN values
plt.title('Histogram of Converted Compensation')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')
plt.grid(True)  # Optional: Adds a grid to the plot
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
QUERY_AGE = """
SELECT Age
FROM master
"""
age_df = pd.read_sql_query(QUERY_AGE, conn)
# Plot the box plot
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.boxplot(age_df['Age'].dropna(), vert=False)  # Drop NaN values
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(True)  # Optional: Adds a grid to the plot
plt.show()


```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the 'Age' and 'WorkWeekHrs' columns data
QUERY = """
SELECT Age, WorkWeekHrs
FROM master
"""
data_df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Drop rows with NaN in either 'Age' or 'WorkWeekHrs'
data_df = data_df.dropna(subset=['Age', 'WorkWeekHrs'])

# Plot the scatter plot
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.scatter(data_df['Age'], data_df['WorkWeekHrs'], alpha=0.5)  # Set transparency
plt.title('Scatter Plot of Age vs Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid(True)  # Optional: Adds a grid to the plot
plt.show()

```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the 'WorkWeekHrs', 'CodeRevHrs', and 'Age' columns data
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""
data_df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()
# Drop rows with NaN in either 'WorkWeekHrs', 'CodeRevHrs', or 'Age'
data_df = data_df.dropna(subset=['WorkWeekHrs', 'CodeRevHrs', 'Age'])

# Create a bubble plot
plt.figure(figsize=(10, 6))  # Set the figure size (optional)
plt.scatter(
    data_df['WorkWeekHrs'],  # X-axis: Work Week Hours
    data_df['CodeRevHrs'],   # Y-axis: Code Review Hours
    s=data_df['Age']*10,     # Bubble size: Age (scaled for better visibility)
    alpha=0.5,               # Transparency of the bubbles
    edgecolor='black'        # Add border to bubbles
)
plt.title('Bubble Plot of Work Week Hours vs Code Review Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid(True)  # Optional: Adds a grid to the plot
plt.show()

```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the top 5 databases respondents wish to learn next year
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(*) as count
FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY count DESC
LIMIT 5
"""

# Execute the query and load the results into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['count'], labels=df['DatabaseDesireNextYear'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Databases Respondents Wish to Learn Next Year')
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get WorkWeekHrs and CodeRevHrs for age group 30 to 35
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 AND 35
"""

# Execute the query and load the results into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Calculate the median WorkWeekHrs and CodeRevHrs for each age
median_df = df.groupby('Age').median()

# Plot a stacked bar chart
median_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Median Work Week Hours and Code Review Hours for Age Group 30 to 35')
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get ConvertedComp for age group 45 to 60
QUERY = """
SELECT Age, ConvertedComp
FROM master
WHERE Age BETWEEN 45 AND 60
"""

# Execute the query and load the results into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Calculate the median ConvertedComp for each age
median_comp_df = df.groupby('Age')['ConvertedComp'].median()

# Plot the median ConvertedComp
plt.figure(figsize=(10, 6))
plt.plot(median_comp_df, marker='o', linestyle='-', color='b')
plt.title('Median Converted Compensation for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.grid(True)
plt.show()


```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt

# Connect to the SQLite database
conn = sqlite3.connect("m4_survey_data.sqlite")

# Query to get the count of each MainBranch
QUERY = """
SELECT MainBranch, COUNT(*) as Count
FROM master
GROUP BY MainBranch
"""

# Execute the query and load the results into a DataFrame
df = pd.read_sql_query(QUERY, conn)

# Close the database connection
conn.close()

# Plot the horizontal bar chart
plt.figure(figsize=(10, 6))
plt.barh(df['MainBranch'], df['Count'], color='skyblue')
plt.title('Distribution of Main Branches')
plt.xlabel('Count')
plt.ylabel('Main Branch')
plt.grid(True, axis='x')  # Add gridlines to the x-axis
plt.show()
```
